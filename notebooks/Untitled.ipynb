```python
from shapely.geometry import Point
from secret import API_KEY
import geopandas as gpd
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
path = 'data/fridge/fridge'
df = pd.read_csv(path + '_.csv', encoding='utf-8')
df.head()
```

```python
df['geometry'] = df.apply(lambda x: Point(x['long'], x['lat']), axis=1)
df = df.drop(['lat', 'long'], 1)
df.head()
```

```python
df.info()
```

```python
df = gpd.GeoDataFrame(df)
```

```python
df['added_time'] = pd.to_datetime(df.added_time)
df['parsed_at'] = pd.to_datetime(df.parsed_at)
```

```python
df.to_file(path + '.gpkg', driver='GPKG')
```
